# Enterprise Retail Sales & Customer Analytics Platform

## Data Cleaning

### Objective
The objective of this notebook is to clean and validate the retail dataset before performing exploratory data analysis and building dashboards.

**Tools Used**
- Python
- Pandas
- NumPy

```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings("ignore")
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 20)
```

```python
# Load Orders sheet
df = pd.read_csv(
    "Dataset.csv",)
df.head()
```

```python
print("Rows :", df.shape[0])
print("Columns :", df.shape[1])
```

```python
df.columns.tolist()
```

```python
df.info()
```

```python
missing = pd.DataFrame({
    "Missing Values": df.isnull().sum(),
    "Percentage": round(df.isnull().mean()*100,2)
})
missing
```

```python
print("Duplicate Rows :", df.duplicated().sum())
```

```python
df = df.drop_duplicates()
print(df.shape)
```

```python
df.dtypes
```

```python
df["Order Date"] = pd.to_datetime(
    df["Order Date"],
    format="%d-%m-%Y",
    errors="coerce"
)

df["Ship Date"] = pd.to_datetime(
    df["Ship Date"],
    format="%d-%m-%Y",
    errors="coerce"
)
```

```python
df[["Order Date","Ship Date"]].dtypes
```

```python
print(df["Order Date"].isna().sum())
print(df["Ship Date"].isna().sum())
```

```python
df.describe()
```

```python
df.describe(include="object")
```

```python
for col in ["Category","Sub-Category","Segment","Region","Ship Mode"]:
    print("="*50)
    print(col)
    print(df[col].unique())
```

```python
loss_orders = df[df["Profit"]<0]
print("Loss Orders :",len(loss_orders))
```

```python
df["Discount"].value_counts().sort_index()
```

```python
df["Shipping Days"] = (
    df["Ship Date"]-df["Order Date"]
).dt.days
```

```python
df[["Order Date","Ship Date","Shipping Days"]].head()
```

```python
print(df.shape)
```

```python
df.to_csv(
    "clean_superstore.csv",
    index=False
)

print("Clean Dataset Exported Successfully")
```
